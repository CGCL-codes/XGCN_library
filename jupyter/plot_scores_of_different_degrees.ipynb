```python
import numpy as np
from matplotlib import pyplot as plt
import os.path as osp
```

```python
save_root = '/media/xreco/DEV/xiran/data/social_and_user_item/model_outputs/figure-www23'
```

```python
dataset = 'livejournal'
Y = np.loadtxt('data/'+dataset+'/scores_of_different_degrees.csv')
y_node2vec = Y[0]
y_pprgo = Y[1]
y_xgcn = Y[2]
```

```python
x = np.arange(10)
```

```python
def set_axis_font_density(ax, size=14):
    ax.tick_params(labelsize=size)
    labels = ax.get_xticklabels() + ax.get_yticklabels()
    for label in labels:
        label.set_fontname('Times New Roman')

color_xgcn = '#FF2442'
color_pprgo = '#55D8C1'
color_node2vec = '#F9D923'

font_legend = {
    'family': 'Times New Roman',
    'weight': 'normal',
    'size': 18,
}
font_axis_label = {
    'family': 'Times New Roman',
    'weight': 'normal',
    'size': 24,
}
linewidth = 1.5
```

```python
fig, axs = plt.subplots(1 , 1, figsize=(5, 4))

ax = axs

ax.plot(x, y_node2vec, 'o-', linewidth=linewidth, markersize=6, label='node2vec', color=color_node2vec)
ax.plot(x, y_pprgo, 'o-', linewidth=linewidth, markersize=6, label='PPRGo', color=color_pprgo)
ax.plot(x, y_xgcn, 'o-', linewidth=linewidth, markersize=6, label='xGCN', color=color_xgcn)

ax.set_xlabel('percentile of degree', font_axis_label)
ax.set_ylabel('recall@100', font_axis_label)
set_axis_font_density(ax, 16)
# ax.legend(loc='lower right')
ax.legend()

# ax0.legend(prop=font_legend, frameon=True, ncol=2, 
#            bbox_to_anchor=(0.6, 1.0)
# )
# ax.legend(prop=font_legend)#, loc='lower right')
ax.legend(prop=font_legend , loc='lower left')


# plt.savefig(osp.join(save_root, title + '.pdf'), 
#             dpi=300,  bbox_inches='tight')
plt.show()
```
