```python
import numpy as np
from matplotlib import pyplot as plt
import os.path as osp
```

```python
def set_axis_font_density(ax, size=14):
    ax.tick_params(labelsize=size)
    labels = ax.get_xticklabels() + ax.get_yticklabels()
    for label in labels:
        label.set_fontname('Times New Roman')
```

```python
color_pprgo = 'g'
color_dnn = 'r'

font_legend = {
    'family': 'Times New Roman',
    'weight': 'normal',
    'size': 18,
}
font_axis_label = {
    'family': 'Times New Roman',
    'weight': 'normal',
    'size': 20,
}
```

```python
dataset = 'amazon-book'
data_root = '../script_local/TK/'
file = osp.join(data_root, 'TK-' + dataset + '.txt')
X = np.loadtxt(file)
K = ['10', '15', '30', 'inf']
T2 = X[0]
T5 = X[1]
T10 = X[2]
```

```python
color_T1 = '#55D8C1' #'#B3E283' #'#95DAC1'
color_T3 = '#FFD371'
color_T10 = '#FD6F96'
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(5, 4)) #, figsize=(16, 4), sharey=False)

# plt.subplots_adjust(
# #     left=0.125,
# #     bottom=0.1, 
# #     right=0.9, 
# #     top=0.9, 
#     wspace=0.1,
# #     hspace=0.35,
# )

ax0 = axes

x = np.arange(len(K))
ax0.plot(x, T2, 'D-', color=color_T1, label='T=2')
ax0.plot(x, T5, 'D-', color=color_T3, label='T=5')
ax0.plot(x, T10, 'D-', color=color_T10, label='T=10')
# ax0.grid(ls='--')
plt.xticks(x, K)

ax0.set_ylabel('Recall@20', font_axis_label)
ax0.set_xlabel('K', font_axis_label)
set_axis_font_density(ax0, 16)

ax0.set_title(dataset, font_axis_label)

# ax0.legend(prop=font_legend, frameon=True, ncol=2, 
#            bbox_to_anchor=(0.6, 1.0)
# )
ax0.legend(prop=font_legend)#, loc='lower right')

plt.savefig(osp.join(data_root, "fig-" + dataset + ".pdf"), bbox_inches='tight')
plt.show()
```
